```python
def graph_edit_distance_2(graph1, graph2, node_substitution_cost=1, node_addition_cost=1, edge_addition_cost=1, node_removal_cost=1, edge_removal_cost=1, edge_substitution_cost=1):
    total_cost = 0

    # 1 - Node substitution
    common_nodes = set(graph1.nodes) & set(graph2.nodes)
    for node in common_nodes:
        if graph1[node] != graph2[node]:
            total_cost += node_substitution_cost

    # 2 - Node addition
    nodes_only_in_graph2 = set(graph2.nodes) - set(graph1.nodes)
    total_cost += len(nodes_only_in_graph2) * node_addition_cost

    # 3 - Edge addition
    common_edges = set(graph1.edges) & set(graph2.edges)
    for edge in common_edges:
        if graph1[edge[0]][edge[1]] != graph2[edge[0]][edge[1]]:
            total_cost += edge_addition_cost

    # 4 - Node removal
    nodes_only_in_graph1 = set(graph1.nodes) - set(graph2.nodes)
    total_cost += len(nodes_only_in_graph1) * node_removal_cost

    # 5 - Edge removal
    edges_only_in_graph1 = set(graph1.edges) - set(graph2.edges)
    total_cost += len(edges_only_in_graph1) * edge_removal_cost

    # 6 - Edge substitution
    for edge in common_edges:
        if graph1[edge[0]][edge[1]] != graph2[edge[0]][edge[1]]:
            total_cost += edge_substitution_cost

    return total_cost

graph_edit_distance_2(G1, G2)
```
